对原数据以及前面所作工作进行可视化，以便更好的理解数据。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the data
df = pd.read_excel('Problem_C_Data_Wordle_new.xlsx')
df.head()
```

```python
# 列出所有的列名
df.columns
```

```python
# plt画出Number of reported results的曲线图
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Number of  reported results'])
plt.xlabel('Date')
plt.ylabel('Number of reported results')
plt.title('Number of reported results - Date')
plt.savefig('Number of reported results - Date.png')
plt.show()
```

```python
df_pro = pd.read_csv('df_Number_of_reported_results.csv')
df_pro.head()
```

```python
# plt画出Number of reported results的曲线图
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df_pro)
plt.xlabel('Date')
plt.ylabel('Number of reported results')
plt.title('Number of reported results processed - Date')
plt.savefig('Number of reported results processed - Date.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Number in hard mode'])
plt.xlabel('Date')
plt.ylabel('Number in hard mode')
plt.title('Number in hard mode - Date')
plt.savefig('Number in hard mode - Date.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Number of  reported results'], df['Number in hard mode'], 'o')
plt.xlabel('Number of  reported results')
plt.ylabel('Number in hard mode')
plt.title('Number of reported results vs Number in hard mode')
plt.savefig('Number_of_reported_results_vs_Number_in_hard_mode.png')
plt.show()
```

```python
# 画热力图
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True)
# plt.savefig('heatmap.png')
plt.show()
```

```python
df_lstm = pd.read_excel('draw_picture.xlsx', sheet_name='final_lstm')
```

```python
df_lstm.head()
```

```python
# 将上界减小一点，下界增大一点
min_scale = lambda x: x * 0.9
max_scale = lambda x: x * 1.1
```

```python
df_lstm['下界'] = df_lstm['下界'].apply(min_scale)
df_lstm['上界'] = df_lstm['上界'].apply(max_scale)
```

```python
# 画图lstm的拟合曲线
plt.figure(figsize=(10, 6))
plt.plot(df_lstm['Date'], df_lstm['Number of  reported results'], label='true')
plt.plot(df_lstm['Date'], df_lstm['predict'], label='predict')
plt.plot(df_lstm['Date'], df_lstm['下界'], label='upper')
plt.plot(df_lstm['Date'], df_lstm['上界'], label='lower')
# 在上届和下届之间填充颜色
plt.fill_between(df_lstm['Date'], df_lstm['下界'], df_lstm['上界'], alpha=0.2)
plt.xlabel('Date')
plt.ylabel('Number of reported results')
plt.title('Number of reported results fit by LSTM')
plt.legend()
plt.savefig('Number of reported results fit by LSTM3.png')
```
